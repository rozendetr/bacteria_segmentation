```python
import pandas as pd
import cv2
from PIL import Image
import numpy as np
import os
import json
```

```python
DIR_TRAIN = r'/media/vladimir/LENOVO/DATA/ITS/train'
```

```python
list_class = os.listdir(DIR_TRAIN)
```

```python
list_coord_obj = []
i = 1
for name_class in list_class:
    list_files= os.listdir(os.path.join(DIR_TRAIN, name_class))
    for json_file in list_files:
        if json_file.endswith(".json"):
            print(os.path.join(name_class, json_file))
            jpg_file = os.path.splitext(json_file)[0]+'.jpg'
            with open(os.path.join(DIR_TRAIN, name_class, json_file), 'r') as j:
                contents = json.loads(j.read())
            width_image = contents['imageWidth']
            height_image = contents['imageHeight']
            shapes = contents['shapes']
            for shape in shapes:
                row = [jpg_file,
                      width_image,
                      height_image,
                      name_class,
                      shape['label'],
                      shape['points'],
                      ]
                list_coord_obj.append(row)
            print(i)
```

```python
headers = ['name_file', 'width', 'height', 'class', 'label', 'points']
pd.DataFrame(list_coord_obj, columns=headers).to_csv('coord_objs.csv', index=False)
```
